```python
import requests
from bs4 import BeautifulSoup
import csv

# List of URLs to scrape
urls = [
    "https://www.transfermarkt.com/fc-zurich/spielplandatum/verein/260/plus/1?saison_id=2019&wettbewerb_id=&day=&heim_gast=&punkte=&datum_von=&datum_bis=", # zurich 19-20
    "https://www.transfermarkt.com/fc-zurich/spielplandatum/verein/260/plus/1?saison_id=2020&wettbewerb_id=&day=&heim_gast=&punkte=&datum_von=&datum_bis=", # 20-21
    "https://www.transfermarkt.com/fc-zurich/spielplandatum/verein/260/plus/1?saison_id=2021&wettbewerb_id=&day=&heim_gast=&punkte=&datum_von=&datum_bis=", # 21-22
    "https://www.transfermarkt.com/fc-zurich/spielplandatum/verein/260/plus/1?saison_id=2022&wettbewerb_id=&day=&heim_gast=&punkte=&datum_von=&datum_bis=", # 22-23
    "https://www.transfermarkt.com/fc-zurich/spielplandatum/verein/260/plus/1?saison_id=2023&wettbewerb_id=&day=&heim_gast=&punkte=&datum_von=&datum_bis=", # 23-24
    "https://www.transfermarkt.com/fc-zurich/spielplandatum/verein/260/plus/1?saison_id=2024&wettbewerb_id=&day=&heim_gast=&punkte=&datum_von=&datum_bis=", # 24-25
    "https://www.transfermarkt.com/fc-lausanne-sport/spielplandatum/verein/527/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # lausanne 19-20
    "https://www.transfermarkt.com/fc-lausanne-sport/spielplandatum/verein/527/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/fc-lausanne-sport/spielplandatum/verein/527/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/fc-lausanne-sport/spielplandatum/verein/527/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/fc-lausanne-sport/spielplandatum/verein/527/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/fc-lausanne-sport/spielplandatum/verein/527/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",# 24-25
    "https://www.transfermarkt.com/servette-fc/spielplandatum/verein/61/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # Geneve 19-20
    "https://www.transfermarkt.com/servette-fc/spielplandatum/verein/61/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/servette-fc/spielplandatum/verein/61/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/servette-fc/spielplandatum/verein/61/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/servette-fc/spielplandatum/verein/61/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/servette-fc/spielplandatum/verein/61/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 24-25
    "https://www.transfermarkt.com/bsc-young-boys/spielplandatum/verein/452/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # BSC Young Boys 19-20
    "https://www.transfermarkt.com/bsc-young-boys/spielplandatum/verein/452/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/bsc-young-boys/spielplandatum/verein/452/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/bsc-young-boys/spielplandatum/verein/452/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/bsc-young-boys/spielplandatum/verein/452/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/bsc-young-boys/spielplandatum/verein/452/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 24-25
    "https://www.transfermarkt.com/fc-lugano/spielplandatum/verein/2790/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 19-20 FC Lugano
    "https://www.transfermarkt.com/fc-lugano/spielplandatum/verein/2790/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/fc-lugano/spielplandatum/verein/2790/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/fc-lugano/spielplandatum/verein/2790/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/fc-lugano/spielplandatum/verein/2790/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/fc-lugano/spielplandatum/verein/2790/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 24-25
    "https://www.transfermarkt.com/fc-st-gallen-1879/spielplandatum/verein/257/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 19-20 FC St.Gallen
    "https://www.transfermarkt.com/fc-st-gallen-1879/spielplandatum/verein/257/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/fc-st-gallen-1879/spielplandatum/verein/257/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/fc-st-gallen-1879/spielplandatum/verein/257/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/fc-st-gallen-1879/spielplandatum/verein/257/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/fc-st-gallen-1879/spielplandatum/verein/257/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 24-25
    "https://www.transfermarkt.com/fc-winterthur/spielplandatum/verein/242/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 19-20 Winterthur
    "https://www.transfermarkt.com/fc-winterthur/spielplandatum/verein/242/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/fc-winterthur/spielplandatum/verein/242/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/fc-winterthur/spielplandatum/verein/242/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/fc-winterthur/spielplandatum/verein/242/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/fc-winterthur/spielplandatum/verein/242/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 24-25
    "https://www.transfermarkt.com/fc-luzern/spielplandatum/verein/434/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 19-20 Luzern
    "https://www.transfermarkt.com/fc-luzern/spielplandatum/verein/434/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/fc-luzern/spielplandatum/verein/434/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/fc-luzern/spielplandatum/verein/434/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/fc-luzern/spielplandatum/verein/434/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/fc-luzern/spielplandatum/verein/434/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 24-25
    "https://www.transfermarkt.com/fc-basel-1893/spielplandatum/verein/26/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 19-20 Basel
    "https://www.transfermarkt.com/fc-basel-1893/spielplandatum/verein/26/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/fc-basel-1893/spielplandatum/verein/26/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/fc-basel-1893/spielplandatum/verein/26/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/fc-basel-1893/spielplandatum/verein/26/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/fc-basel-1893/spielplandatum/verein/26/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 24-25
    "https://www.transfermarkt.com/fc-sion/spielplandatum/verein/321/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 19-20 Sion
    "https://www.transfermarkt.com/fc-sion/spielplandatum/verein/321/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/fc-sion/spielplandatum/verein/321/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/fc-sion/spielplandatum/verein/321/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/fc-sion/spielplandatum/verein/321/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/fc-sion/spielplandatum/verein/321/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 24-25
    "https://www.transfermarkt.com/yverdon-sport-fc/spielplandatum/verein/322/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 19-20 Yverdon
    "https://www.transfermarkt.com/yverdon-sport-fc/spielplandatum/verein/322/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/yverdon-sport-fc/spielplandatum/verein/322/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/yverdon-sport-fc/spielplandatum/verein/322/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/yverdon-sport-fc/spielplandatum/verein/322/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/yverdon-sport-fc/spielplandatum/verein/322/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 24-25
    "https://www.transfermarkt.com/grasshopper-club-zurich/spielplandatum/verein/504/saison_id/2019/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 19-20 Grasshoper club Zurich
    "https://www.transfermarkt.com/grasshopper-club-zurich/spielplandatum/verein/504/saison_id/2020/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 20-21
    "https://www.transfermarkt.com/grasshopper-club-zurich/spielplandatum/verein/504/saison_id/2021/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 21-22
    "https://www.transfermarkt.com/grasshopper-club-zurich/spielplandatum/verein/504/saison_id/2022/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 22-23
    "https://www.transfermarkt.com/grasshopper-club-zurich/spielplandatum/verein/504/saison_id/2023/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 23-24
    "https://www.transfermarkt.com/grasshopper-club-zurich/spielplandatum/verein/504/saison_id/2024/wettbewerb_id//datum_von/0000-00-00/datum_bis/0000-00-00/day/0/plus/1",  # 24-25
]

# Headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# CSV file columns
columns = ["Competition", "Matchday", "Date", "Time", "Home Team", "Ranking Home Team", "Away Team", "Ranking Away Team", "Attendance", "Result"]

# Create a list to store all extracted data
all_data = []

# Loop through each URL
for url in urls:
    print(f"Scraping: {url}")
    # Send a GET request
    response = requests.get(url, headers=headers)
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the table within the responsive-table div
    table = soup.find("div", class_="responsive-table").find("table")
    # Extract all rows from tbody
    rows = table.find("tbody").find_all("tr")

    # Initialize a variable to hold the current competition name
    current_competition = ""

    # Loop through each row
    for row in rows:
        # Check if the row contains a competition name
        competition_cell = row.find("td", class_="extrarow bg_blau_20 hauptlink")
        if competition_cell:
            # Update the current competition name
            current_competition = competition_cell.get_text(strip=True)
            continue  # Skip to the next row since this row doesn't contain match data

        cells = row.find_all("td")  # Extract all cells within the row
        if len(cells) >= 11:  # Ensure the row contains enough data columns
            matchday = cells[0].get_text(strip=True)  # First column: Matchday
            date = cells[1].get_text(strip=True)      # Second column: Date
            time = cells[2].get_text(strip=True)      # Third column: Time

            # Extract home team and its ranking
            home_team_cell = cells[4]
            home_team = home_team_cell.find("a").get_text(strip=True)  # Extract team name
            home_ranking_span = home_team_cell.find("span", class_="tabellenplatz")
            home_ranking = home_ranking_span.get_text(strip=True) if home_ranking_span else ""  # Extract ranking

            # Extract away team and its ranking
            away_team_cell = cells[6]
            away_team = away_team_cell.find("a").get_text(strip=True)  # Extract team name
            away_ranking_span = away_team_cell.find("span", class_="tabellenplatz")
            away_ranking = away_ranking_span.get_text(strip=True) if away_ranking_span else ""  # Extract ranking

            attendance = cells[9].get_text(strip=True)  # Attendance
            result = cells[10].get_text(strip=True)     # Result

            # Append the row data to the list
            all_data.append([current_competition, matchday, date, time, home_team, home_ranking, away_team, away_ranking, attendance, result])

# Write the data to a CSV file
csv_file = "transfermarkt_data_with_competitions.csv"
with open(csv_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(columns)  # Write the header
    writer.writerows(all_data)  # Write all rows

print(f"Data successfully written to {csv_file}")
```

```python
import csv
import requests
from datetime import datetime
from google.colab import files
import os
import time

# Dictionary mapping club names to their coordinates
club_coordinates = {
    "FC Z체rich": {"lat": 47.3820, "lon": 8.5046},
    "FC Basel": {"lat": 47.5417, "lon": 7.6203},
    "BSC Young Boys": {"lat": 46.9632, "lon": 7.4648},
    "FC Lugano": {"lat": 46.0222, "lon": 8.9608},
    "FC Luzern": {"lat": 47.0333, "lon": 8.3052},
    "Servette FC": {"lat": 46.1778, "lon": 6.1275},
    "Grasshoppers": {"lat": 47.3820, "lon": 8.5046},
    "Yverdon Sport": {"lat": 46.7833, "lon": 6.6471},
    "Lausanne-Sport": {"lat": 46.5411, "lon": 6.6094},
    "FC Sion": {"lat": 46.2339, "lon": 7.3611},
    "FC Winterthur": {"lat": 47.4992, "lon": 8.7375},
    "FC St. Gallen": {"lat": 47.4083, "lon": 9.3103},
}

# Function to fetch weather data using Open Meteo Historical API

def get_weather_open_meteo(date, time, home_team):
    try:
        # Get coordinates for the home team
        coordinates = club_coordinates.get(home_team)
        if not coordinates:
            return "None", "None"  # Fallback if the team is not in the dictionary

        lat = coordinates["lat"]
        lon = coordinates["lon"]

        # Format date to YYYY-MM-DD and extract hour
        try:
            dt_object = datetime.strptime(f"{date} {time}", "%a %m/%d/%y %I:%M %p")
            formatted_date = dt_object.strftime("%Y-%m-%d")
            match_hour = dt_object.hour
        except ValueError as ve:
            print(f"Error parsing date/time for row: date='{date}', time='{time}' - {ve}")
            return "None", "None"

        # Check if the match date is in the future
        current_date = datetime.now()
        if dt_object > current_date:
            return "None", "None"

        # Call Open Meteo Historical Weather API
        weather_url = f"https://archive-api.open-meteo.com/v1/archive"
        params = {
            "latitude": lat,
            "longitude": lon,
            "start_date": formatted_date,
            "end_date": formatted_date,
            "hourly": "temperature_2m,weathercode",
            "timezone": "Europe/Zurich"
        }

        while True:
            response = requests.get(weather_url, params=params)
            if response.status_code == 200:
                weather_data = response.json()
                hourly_data = weather_data.get("hourly", {})
                temperatures = hourly_data.get("temperature_2m", [None])
                weather_codes = hourly_data.get("weathercode", [None])

                # Adjust to ensure correct hour is within bounds of returned data
                if len(temperatures) > match_hour and len(weather_codes) > match_hour:
                    temperature = temperatures[match_hour]
                    weather_code = weather_codes[match_hour]

                    # Translate weather code to descriptive text
                    weather_description = "Unknown"
                    if weather_code is not None:
                        if weather_code in [0, 1]:
                            weather_description = "Clear or mostly clear"
                        elif weather_code in [2, 3]:
                            weather_description = "Partly cloudy"
                        elif weather_code in [45, 48]:
                            weather_description = "Foggy"
                        elif weather_code in [51, 53, 55, 56, 57]:
                            weather_description = "Drizzle"
                        elif weather_code in [61, 63, 65, 66, 67]:
                            weather_description = "Rainy"
                        elif weather_code in [71, 73, 75, 77]:
                            weather_description = "Snowy"
                        elif weather_code in [80, 81, 82]:
                            weather_description = "Showers"
                        elif weather_code in [95, 96, 99]:
                            weather_description = "Thunderstorm"

                    weather = weather_description
                else:
                    weather = "None"
                    temperature = "None"

                return weather, temperature
            elif response.status_code == 429:
                print("API request limit exceeded. Retrying in 60 seconds...")
                time.sleep(60)
            else:
                print(f"API request failed with status code {response.status_code} and response: {response.text}")
                return "None", "None"
    except Exception as e:
        print(f"Error fetching weather using Open Meteo API: {e}")
        return "None", "None"

# Read the existing CSV file
input_csv = "transfermarkt_data_with_competitions.csv"  # Input CSV
output_csv = "transfermarkt_data_with_competitions_and_weather_next_200_part12.csv"  # Output CSV for next 200 rows

# Process the next 200 rows (rows 2400 to 2600)
start_row = 2400
end_row = 2600

with open(input_csv, mode="r", encoding="utf-8-sig") as infile:
    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames + ["Weather", "Temperature (째C)"]  # Add new columns
    rows = list(reader)

    with open(output_csv, mode="w", newline="", encoding="utf-8-sig") as outfile:
        writer = csv.DictWriter(outfile, fieldnames=fieldnames)

        # Write the header
        writer.writeheader()

        for row in rows[start_row:end_row]:
            # Extract match details
            date = row.get("Date")
            time = row.get("Time")
            home_team = row.get("Home Team")  # Match with the club name

            if not date or not time or not home_team:
                row["Weather"] = "None"
                row["Temperature (째C)"] = "None"
            else:
                # Fetch weather data using Open Meteo API
                weather, temperature = get_weather_open_meteo(date, time, home_team)

                # Append weather data to the row
                row["Weather"] = weather
                row["Temperature (째C)"] = temperature

            # Write the updated row to the output CSV
            writer.writerow(row)

print(f"Weather data successfully added to {output_csv} for rows 2400 to 2599.")

# Download the final CSV file to the user's computer
files.download(output_csv)
```

```python
import csv
from google.colab import files

# List of input CSV files
input_csv_files = [
    "transfermarkt_data_with_competitions_and_weather_first_200.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part2.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part3.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part4.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part5.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part6.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part7.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part8.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part9.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part10.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part11.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part12.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part13.csv",
    "transfermarkt_data_with_competitions_and_weather_next_200_part14.csv"
]

# Output CSV file
output_csv = "transfermarkt_data_with_competitions_and_weather_complete.csv"

# Combine all CSV files into one
with open(output_csv, mode="w", newline="", encoding="utf-8-sig") as outfile:
    writer = None
    for input_csv in input_csv_files:
        with open(input_csv, mode="r", encoding="utf-8-sig") as infile:
            reader = csv.DictReader(infile)
            if writer is None:
                # Write the header only once
                writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)
                writer.writeheader()
            for row in reader:
                writer.writerow(row)

print(f"All datasets successfully combined into {output_csv}.")

# Download the combined CSV file to the user's computer
files.download(output_csv)
```
